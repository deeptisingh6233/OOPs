```python
import json
import os


class Book:

    def __init__(self, book_id, title, author, quantity):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.quantity = quantity

    def to_dict(self):
        return {
            "book_id": self.book_id,
            "title": self.title,
            "author": self.author,
            "quantity": self.quantity
        }


class Library:

    def __init__(self):
        self.file_name = "books.json"

    def load_books(self):

        if not os.path.exists(self.file_name):
            return []

        with open(self.file_name, "r") as file:
            try:
                return json.load(file)
            except:
                return []

    def save_books(self, books):

        with open(self.file_name, "w") as file:
            json.dump(books, file, indent=4)

    def add_book(self):

        try:
            book_id = int(input("Enter Book ID: "))
            title = input("Enter Title: ")
            author = input("Enter Author: ")
            quantity = int(input("Enter Quantity: "))

            books = self.load_books()

            for book in books:
                if book["book_id"] == book_id:
                    print("Book ID already exists")
                    return

            new_book = Book(
                book_id,
                title,
                author,
                quantity
            )

            books.append(new_book.to_dict())

            self.save_books(books)

            print("Book added successfully")

        except ValueError:
            print("Invalid input")

    def show_books(self):

        books = self.load_books()

        if not books:
            print("No books available")
            return

        print("\n------ BOOKS ------")

        for book in books:

            print(
                f"ID: {book['book_id']} | "
                f"Title: {book['title']} | "
                f"Author: {book['author']} | "
                f"Quantity: {book['quantity']}"
            )

    def search_book(self):

        title = input("Enter title to search: ")

        books = self.load_books()

        found = False

        for book in books:

            if title.lower() in book["title"].lower():

                print(book)

                found = True

        if not found:
            print("Book not found")

    def remove_book(self):

        book_id = int(input("Enter Book ID: "))

        books = self.load_books()

        updated_books = []

        found = False

        for book in books:

            if book["book_id"] == book_id:
                found = True

            else:
                updated_books.append(book)

        if found:

            self.save_books(updated_books)

            print("Book removed")

        else:

            print("Book not found")

    def update_quantity(self):

        book_id = int(input("Enter Book ID: "))
        amount = int(input("Enter Quantity To Add: "))

        books = self.load_books()

        found = False

        for book in books:

            if book["book_id"] == book_id:

                book["quantity"] += amount

                found = True

        if found:

            self.save_books(books)

            print("Quantity Updated")

        else:

            print("Book not found")

    def menu(self):

        while True:

            print("""
1. Add Book
2. Show Books
3. Search Book
4. Remove Book
5. Update Quantity
6. Exit
""")

            try:

                choice = int(input("Enter choice: "))

                if choice == 1:
                    self.add_book()

                elif choice == 2:
                    self.show_books()

                elif choice == 3:
                    self.search_book()

                elif choice == 4:
                    self.remove_book()

                elif choice == 5:
                    self.update_quantity()

                elif choice == 6:
                    print("Thank You")
                    break

                else:
                    print("Invalid Choice")

            except ValueError:
                print("Enter valid number")


library = Library()
library.menu()
```

```python
class Car(Vehicle):
  def __init__(self,vehicle_number):
    self.vehicle_number=vehicle_number
    super().__init__(vehicle_number)

  def calculate_fare(self,distance):
    return distance*20

class Auto(Vehicle):
  def __init__(self,vehicle_number):
    self.vehicle_number=vehicle_number
    super().__init__(vehicle_number)

  def calculate_fare(self,distance):
    return distance*10

class Bike(Vehicle):
  def __init__(self,vehicle_number):
    self.vehicle_number=vehicle_number
    super().__init__(vehicle_number)

  def calculate_fare(self,distance):
    return distance*15
```

```python
class Payment:

    def __init__(self, amount, payment_method):
        self.amount = amount
        self.payment_method = payment_method
        self.status = "Pending"

    def make_payment(self):
        self.status = "Paid"
        print(f"₹{self.amount} paid successfully via {self.payment_method}")

    def view_payment(self):
        print(f"""
Amount : ₹{self.amount}
Method : {self.payment_method}
Status : {self.status}
""")
```

```python
bike = Bike("UP32AB1234")

fare = bike.calculate_fare(10)

payment = Payment(fare, "PayTm")

payment.make_payment()
payment.view_payment()
```
